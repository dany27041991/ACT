# French-DeGroot opinion pooling

## DeGroot Learning Model: Premessa
Il modello di apprendimento DeGroot è un modello semplice e importante di come le persone in una rete aggiornano le loro opinioni nel tempo e alla fine raggiungono un consenso di gruppo. E' stato uno dei primi modelli di formazione delle opinioni basato su agenti 4 ed è stato proposto dallo psicologo sociale French nel suo influente articolo ( French, 1956 ), che collega insieme la SNA e la teoria dei sistemi. Insieme alla sua generalizzazione, suggerita da DeGroot (1974) e chiamata “iterative opinion pooling”, questo modello descrive una procedura semplice, che consente a diversi agenti razionali di raggiungere il consenso (Lehrer, 1976; Lehrer & Wagner, 1981; Wagner, 1978 ); può anche essere considerato come un algoritmo di apprendimento non bayesiano (Golub & Jackson, 2010; Jadbabaie, Molavi, Sandroni e Tahbaz-Salehi, 2012). L’obiettivo originale di French, tuttavia, non era quello di studiare il consenso e i meccanismi di apprendimento, ma piuttosto di trovare un modello matematico per il potere sociale (French & Raven, 1959; French, 1956; Friedkin, 1986). Il potere sociale di un individuo nel gruppo è la sua capacità di controllare il comportamento del gruppo, indicando così la centralità del nodo dell'individuo nella rete sociale. Il lavoro di French ha quindi rivelato una profonda relazione tra formazione dell’opinione e misure di centralità.

### Idea di base
In ogni periodo, ogni agente nella rete aggiorna la sua opinione prendendo una media ponderata della propria opinione e di quelle dei suoi vicini, con pesi costanti e invarianti nel tempo. Date le opinioni iniziali di tutti gli agenti e la matrice che descrive il peso (costante) che ciascun agente attribuisce all’opinione di ogni altro agente, saremo in grado di calcolare la dinamica delle opinioni di tutti e la convinzione di consenso del gruppo a lungo termine. Tra gli altri risultati interessanti, vedremo che l’influenza di un agente sulla convinzione consensuale di lungo periodo del gruppo è determinata dalla centralità del suo autovettore nella matrice dei pesi.

## Ha senso la media ripetuta? 
Prima di entrare nel merito, facciamo un passo indietro e chiediamoci cosa può e non può catturare un modello di media ripetuta. Innanzitutto, comprendere l’apprendimento sociale – ovvero il modo in cui gruppi di persone aggregano le proprie informazioni e formano credenze e preferenze – è una questione chiave nelle scienze sociali, con innumerevoli applicazioni. 
* Quale candidato diventa popolare e vince le elezioni? 
* Quali sono i prezzi delle azioni delle varie società sul mercato azionario?
* Quali ristoranti, libri o film diventano popolari?

Buoni modelli di apprendimento sociale possono guidare l’analisi empirica, le decisioni e le politiche su questi argomenti. Quindi il problema generale è piuttosto importante. 

Esistono due approcci generali per modellare l’apprendimento sociale. 

#### Apprendimento sociale bayesiano: 
Gli agenti sono statistici bayesiani. Gli agenti stanno cercando di apprendere uno stato del mondo sconosciuto (ad esempio, la redditività a lungo termine di un'azienda), iniziano con una convinzione precedente e la aggiornano utilizzando la regola di Bayes alla luce delle informazioni che ricevono dall'ambiente o da altri agenti . 
* Questi modelli rappresentano in un certo senso il “gold standard” per comprendere l’apprendimento razionale. Tuttavia, presuppongono molta razionalità da parte degli agenti e possono essere molto complicati al di fuori dei semplici esempi.
* Grandi modelli di come le persone dovrebbero imparare, se siano grandi modelli di come le persone imparano è più sfumato.

#### Apprendimento sociale non bayesiano: 
Gli agenti fanno qualcosa di più semplice, più “euristico” dell’apprendimento bayesiano. Potresti forse pensare a questi modelli come a
* un'approssimazione approssimativa dell'apprendimento bayesiano quando è troppo complicato, o
* a un modello più accurato per persone reali che non sono così razionali, o
* una descrizione dell'aggiornamento di qualcosa che non è proprio una "credenza" nel vero senso probabilistico (ad esempio, come le persone aggiornano la loro "opinione" su un problema o i loro gusti per un prodotto o un altro).

Queste varie interpretazioni possono avere più o meno senso in contesti diversi.

Il modello di apprendimento DeGroot è il modello più semplice e conosciuto di apprendimento sociale non bayesiano in una rete.
* La procedura di media ripetuta di DeGroot non è tipicamente ben fondata in termini di apprendimento sociale razionale. (Gli agenti nel modello non sono “buoni bayesiani”.)
* Tuttavia, il modello è ancora utile e importante per le ragioni sopra esposte.
* È utile anche come complemento ai modelli bayesiani, che probabilmente hanno basi più solide ma sono spesso meno trattabili.

## Recap Modelli bayesiani
I modelli bayesiani sono modelli statistici che si basano sul teorema di Bayes per inferire delle distribuzioni di probabilità su parametri di interesse a partire dai dati osservati. Questi modelli sono particolarmente utili quando si ha a che fare con l'incertezza nei dati o nei parametri del modello stesso.

Il teorema di Bayes afferma che la distribuzione di probabilità di un parametro dato un insieme di dati può essere calcolata a partire dalla distribuzione a priori del parametro e dalla verosimiglianza dei dati dato il parametro. In breve, il teorema può essere espresso come:

$$
P(\theta | D) = \frac{P(D | \theta) \cdot P(\theta)}{P(D)}
$$

dove:
- $P(\theta | D)$ è la distribuzione di probabilità a posteriori del parametro $\theta$ dato l'insieme di dati D,
- $P(D | \theta)$ è la verosimiglianza dei dati D dato il parametro $\theta$,
- $P(\theta)$ è la distribuzione di probabilità a priori del parametro $\theta$,
- $P(D)$ è la probabilità marginale dei dati, che può essere interpretata come un fattore di normalizzazione.

I modelli bayesiani consentono di incorporare conoscenze a priori sui parametri del modello, consentendo una maggiore flessibilità nella gestione dell'incertezza e nell'aggiornamento delle stime dei parametri in base ai dati osservati. Questa caratteristica li rende particolarmente utili in una vasta gamma di applicazioni, dall'analisi dei dati al machine learning, dall'ingegneria alla medicina e molto altro ancora.

## Il modello French-DeGroot
Il modello French-DeGroot descrive un processo di formazione delle opinioni in un tempo discreto in un gruppo di n agenti, le cui opinioni da quel momento in poi saranno denotate con $x_1 , . . . , x_n$. Consideriamo innanzitutto il caso delle opinioni scalari $x_i∈R$ . Il parametro chiave del modello è una matrice stocastica $n×n$ di pesi di influenza $W=(w_{ij})$. I pesi di influenza $w_{ij}≥0$, dove $j=1,...,n$ possono essere considerati come una risorsa finita, distribuita dall'agente i a se stesso e agli altri agenti. Dato un peso di influenza positivo $w_{ij}>0$, l'agente j è in grado di influenzare l'opinione dell'agente i in ogni fase dell'iterazione dell'opinione; maggiore è il peso assegnato all'agente j, maggiore è la sua influenza sull'agente i. 

![title](./images/french-degroot_example.png)

Matematicamente, il vettore di opinioni $x(k)=(x_1(k),...,x_n(k))^T$ obbedisce all'equazione
$$Equazione_1: x(k+1)=Wx(k), k=0,1,.... $$
che è equivalente al sistema di equazioni
$$Equazione_2: x_i(k+1)=\sum_{j=1}^n w_{ij} x_j(k) $$

Quindi $w_{ij}$ è il contributo dell’opinione dell’agente j in ogni passo dell’iterazione dell’opinione all’opinione dell’agente i nel passo successivo. Il peso dell'autoinfluenza $w_{ii}≥0$ indica l'apertura dell'agente all'assimilazione delle opinioni degli altri: l'agente con $w_{ii}=0$ ha una mentalità aperta e si affida completamente alle opinioni degli altri, mentre l'agente con $w_{ii}=1$ (e $w_{ij}=0 ∀ j= i$) è un agente testardo, “ancorato” alla sua opinione iniziale $x_i(k)≡x_i(0)$. Più in generale, le opinioni dell’agente possono essere vettori di dimensione m, opportunamente rappresentati dalle righe $x_i=(x_i 1,...,x_{im})$. Sovrapponendo queste righe una sull'altra si ottiene una matrice di opinioni $X=(x_{il}) ∈ R n×m$ . L'$Equazione_1$ dovrebbe essere sostituita da
$$Modello French-DeGroot: X(k+1)=WX(k), k=0,1,....$$

Ogni colonna $x_i(k)=(x_{1i}(k),...,x_{ni}(k))^T∈R^n$ di $X(k)$, ovviamente, evolve secondo l'$Equazione_1$. D'ora in poi il modello precedente con matrice stocastica generale W verrà denominato modello French-DeGroot.

### Convergenza
Il modello convergerà ad un consenso condiviso se per qualunque condizione iniziale $x(0)$ esiste il limite
$$ x(∞)=\lim_{k→∞}x(k) = \lim_{k→∞} W^kx(0). $$
Un modello convergente raggiunge il consenso se $x_1(∞) =... = x_n(∞)$ per ogni vettore di opinione iniziale $x(0)$. La convergenza e il consenso nel modello $Equazione_1$ equivalgono, rispettivamente, alla regolarità e alla piena regolarità della matrice stocastica W. In sintesi stiamo dicendo che: la piena regolarità di una matrice stocastica W si riferisce al fatto che la matrice ha tutte le sue righe e colonne somma pari a 1 e tutti i suoi elementi sono non negativi.

### Stubborn Agents
Sebbene il consenso sia un comportamento tipico del modello (2), ci sono situazioni in cui le opinioni non raggiungono il consenso ma si dividono in diversi cluster. Uno dei motivi è la presenza di agenti ostinati (chiamati anche radicali (Hegselmann & Krause, 2015).
Un agente si dice testardo se la sua opinione rimane invariata indipendentemente da quelle degli altri. 

![title](./images/french_degroot_stubborn_example.png)

Se le opinioni obbediscono al modello $Equazione_1$ allora l'agente i è testardo $x_i(k)≡x_i(0)$ se e solo se $w_{ii}=1$.

#### Teorema
Il modello $Equazione_1$ è convergente se e solo se tutte le componenti forti chiuse in G[ W ] sono aperiodiche. Il modello $Equazione_1$ raggiunge il consenso se e solo se G[W] è quasi-fortemente connesso e l'unica componente forte chiusa è aperiodica.

Un tale agente corrisponde ad un nodo sorgente in un grafo G[W], cioè un nodo che non ha archi entranti tranne che per il self-loop.
Questo teorema implica che se G[W] ha l'unica sorgente, essendo anche un nodo root, allora le opinioni raggiungono un consenso (il nodo sorgente è l'unica componente forte chiusa del grafo). Se esiste più di un agente ostinato (cioè G[W] ha diverse fonti), allora il consenso tra di loro è, ovviamente, impossibile. Il Teorema implica, tuttavia, che tipicamente le opinioni in un tale gruppo convergono. 
* Supponiamo che il gruppo abbia $s≥1$ agenti testardi, che influenzano tutti gli altri individui (cioè l'insieme dei nodi sorgente è collegato tramite passeggiate a tutti gli altri nodi di G[W]). Allora il modello è convergente. Infatti, i nodi sorgente sono le uniche componenti forti chiuse di G[W] , che sono ovviamente aperiodiche.

```python
#import delle librerie numpy e matplotlib
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(precision=15)
```

```python
#definisco la funzione che calcola l'average consensus
def compute_average_consensus(W, x0, num_iterations):
    #lim per k -> infinito di W^k * x(0)
    W_k = np.linalg.matrix_power(W, num_iterations*1000)
    result = np.dot(W_k, x0)
    print("POSSIBLE AVERAGE CONSENSUS: ")
    print(result)
    return result
```

```python
#definisco la funzione per il french degroot model
def french_degroot_model(W, x0, num_iterations):
    """
     Modello di consenso French-DeGroot.

     parametri:
         W (numpy.ndarray): matrice di adiacenza che rappresenta l'influenza di un agente rispetto ad un altro. 
             La forma della matrice dovrebbe essere (n, n) dove n è il numero di agenti. 
             La voce (i, j) rappresenta l'influenza dell'agente j sull'agente i.
             I valori devono essere compresi tra 0 e 1 e la somma di ciascuna colonna deve essere 1.
         x0 (numpy.ndarray): opinioni iniziali degli agenti. Dovrebbe essere una matrice 1D di lunghezza n, dove n è il numero di agenti.
         num_iterations (int): numero di iterazioni per eseguire il modello.

     Ritorna:
         numpy.ndarray: opinioni di consenso finale dopo il numero specificato di iterazioni.
         numpy.ndarray: opinioni di ciascun agente ad ogni iterazione. Forma (num_iterazioni, n).
     """

    n = len(x0)
    opinions = x0.copy()
    opinions_history = [opinions.copy()]

    for _ in range(num_iterations):
        opinions = np.dot(W, opinions)
        opinions_history.append(opinions.copy())

    return opinions, np.array(opinions_history)
```

# ESEMPIO 1 - CONVERGENZA SENZA STUBBORN

```python
# Definizione della matrice di adiacenza (rappresenta l'influenza tra agenti)
W = np.array([[1/2, 1/2, 0],
              [1/3, 1/3, 1/3],
              [0, 1/2, 1/2]])

# Condizioni iniziali degli agenti
x0 = np.array([1/2, 1/3, 1/2]).reshape(-1, 1)

# Numero di iterazioni
num_iterations = 10

# Otteniamo le opinioni consensuali finali ed eseguiamo la cronologia delle opinioni
final_opinions, opinions_history = french_degroot_model(W, x0, num_iterations)

# Plotting
plt.figure(figsize=(10, 6))

# Plot delle opinioni di ogni agente
for i in range(len(x0)):
    plt.plot(opinions_history[:, i], label=f'Agent {i+1}')

# Calcoliamo e plottiamo l'average consensus
average_consensus = compute_average_consensus(W, x0, num_iterations)
#plt.plot(np.tile(np.mean(average_consensus), (num_iterations, 1)), label='Average Consensus', linestyle='--', color='black')

average_opinions = np.mean(opinions_history, axis=1)
plt.plot(average_opinions, label='Average Consensus', linestyle='--', color='black')

plt.xlabel('Iterations')
plt.ylabel('Opinions')
plt.title('Evolution of Opinions over Iterations')
plt.legend()
plt.grid(True)
plt.show()
```
